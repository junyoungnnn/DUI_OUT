```python
import pandas as pd
from tensorflow.keras.models import load_model
from sklearn.preprocessing import StandardScaler

# 1. 저장된 모델 로드
model = load_model('best_model.keras')
print("저장된 모델이 로드되었습니다.")

# 2. 21~23년도 데이터로 MinMaxScaler 학습 (예: X_train 사용)
# 학습 데이터 로드 및 스케일러 학습
# 예: X_train = pd.read_csv('train_data.csv')[['자동차 등록 대수', '인구수', '횡단보도 개수', '음식점 수', '버스정류장 개수', '음주 교통사고 건수']]
X_train = pd.read_excel('2024예측데이터셋.xlsx')[['자동차 등록 대수', '인구수', '횡단보도 개수', '음식점 수', '버스정류장 개수', '음주 교통사고 건수']]

scaler = StandardScaler()
scaler.fit(X_train)  # 학습 데이터를 기반으로 StandardScaler 학습
print("StandardScaler가 학습되었습니다.")

# 3. 24년도 데이터 로드
df = pd.read_excel('2024예측데이터셋.xlsx')

# 4. 입력 데이터 준비 및 스케일링
X_2024 = df[['자동차 등록 대수', '인구수', '횡단보도 개수', '음식점 수', '버스정류장 개수', '음주 교통사고 건수']]
X_2024_scaled = scaler.transform(X_2024)  # 스케일러를 사용해 24년도 데이터 변환

# 5. y값 예측
y_2024_pred = model.predict(X_2024_scaled)

# 6. 결과를 DataFrame에 추가
df['예측 음주운전 적발 건수'] = y_2024_pred

# 7. 예측 결과 저장
df.to_csv('24년도_예측결과_deep.csv', index=False, encoding='cp949')
print("24년도 예측 결과가 '24년도_예측결과_deep.csv'로 저장되었습니다.")
```

```python
import xgboost as xgb
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd

# 1. Load the saved XGBoost model
from joblib import load
xgb_loaded = load('xgb_model.joblib')
print("Saved XGBoost model has been loaded.")

# 2. Load training data for scaler fitting (21-23 year data)
X_train = pd.read_excel('2024예측데이터셋.xlsx')[['자동차 등록 대수', '인구수', '횡단보도 개수', '음식점 수', '버스정류장 개수', '음주 교통사고 건수']]

# Fit StandardScaler to training data
scaler = StandardScaler()
scaler.fit(X_train)
print("StandardScaler has been fitted to training data.")

# 3. Load 2024 data for prediction
df = pd.read_excel('2024예측데이터셋.xlsx')

# 4. Prepare input features and scale them
X_2024 = df[['자동차 등록 대수', '인구수', '횡단보도 개수', '음식점 수', '버스정류장 개수', '음주 교통사고 건수']]
X_2024_scaled = scaler.transform(X_2024)

# 5. Predict y values using the loaded model
dmatrix_2024 = xgb.DMatrix(X_2024_scaled)  # Convert to DMatrix
y_2024_pred = xgb_loaded.predict(dmatrix_2024)

# 6. Add predictions to the DataFrame
df['예측 음주운전 적발 건수'] = y_2024_pred

# 7. Save the prediction results
df.to_csv('24년도_예측결과xgb.csv', index=False, encoding='cp949')
print("Prediction results for 2024 have been saved to '24년도_예측결과xgb.csv'.")

```
