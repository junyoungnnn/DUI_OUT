```python
df = pd.read_csv('행정동_음주운전_합본_cp949.csv')
```

```python
df = pd.get_dummies(data=df, columns=['읍면동'], drop_first=True)
```

```python
from sklearn.model_selection import train_test_split

x = df.drop(columns='음주운전 적발 건수', axis=1)
y = df['음주운전 적발 건수']

x_train, x_valid, y_train, y_valid = train_test_split(x, y, test_size=.2, random_state=29)
```

```python
from xgboost import XGBRegressor

xgb = XGBRegressor(max_depth=50, n_estimators=10, random_state=29)
xgb.fit(x_train, y_train)

y_pred = xgb.predict(x_valid)

from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

xgb_mse = mean_squared_error(y_valid, y_pred)
xgb_mae = mean_absolute_error(y_valid, y_pred)
xgb_r2 = r2_score(y_valid, y_pred)
print('MSE:', xgb_mse)
print('MAE:', xgb_mae)
print('r2:', xgb_r2)
print('score:', xgb.score(x_valid, y_valid))
```
