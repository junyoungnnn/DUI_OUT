```python
import pandas as pd
```

```python
df = pd.read_csv('./merged_sheets_cp949.csv', encoding='cp949')
df
```

```python
# 날짜를 datetime 형식으로 변환
df['날짜'] = pd.to_datetime(df['날짜'])

# 연도와 월 컬럼 추가
df['연도'] = df['날짜'].dt.year
df['월'] = df['날짜'].dt.month

# 2021, 2022, 2023 년도 데이터 필터링 (추가 데이터가 있을 경우 확장 가능)
filtered_df = df[df['연도'].isin([2021, 2022, 2023])]

# 구별, 연도별, 월별 음주운전 단속 건수 평균 계산
result = filtered_df.groupby(['부산구별', '연도'])['음주운전 단속건수'].mean().reset_index()

# 결과를 원하는 형식으로 변환
def format_output(row):
    return f"{row['부산구별']} {row['연도']} {row['음주운전 단속건수']}건"

result['결과'] = result.apply(format_output, axis=1)

# 결과 출력
for row in result['결과']:
    print(row)
```

```python
# 결과를 CSV 파일로 저장
result = result.drop(columns='결과')
result.to_csv("monthly_dui_analysis.csv", index=False, encoding='cp949')
```
