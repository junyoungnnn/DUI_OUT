```python
import pandas as pd
import numpy as np
```

```python
path = ['./강서구/', './금정구/', './기장군/', './남구/', './동구/', './동래구/', './부산진구/', './북구/', './사상구/', './사하구/', './서구/', './수영구/', './연제구/', './영도구/', './중구/', './해운대구/']
dong = ['강동동', '서제1동', '기장읍', '대연제1동', '초량제1동', '수민동', '부전제1동', '구포제1동', '삼락동', '괴정제1동', '동대신제1동', '남천제1동', '거제제1동', '남항동', '중앙동', '우제1동']

for j in range(len(path)):
    for i in range(3):
        year = ['2021', '2022', '2023']
        file = f'{dong[j]}{year[i]}.csv'
        name = f'{dong[j]}groupby{year[i]}'
        
        df = pd.read_csv(f'{path[j]}{file}')
    
        # year, month별로 value 평균
        grouped_df = df.groupby(['year', 'month'])['value'].mean().reset_index()

        # value를 소수점 한 자리로 반올림
        grouped_df['value'] = grouped_df['value'].round(1)
        
        # 결과 확인
        print(grouped_df)
        
        # 새로운 CSV로 저장
        output_path = f"{name}.csv"
        grouped_df.to_csv(f'{path[j]}{output_path}', index=False)
```
