```python
import pandas as pd
import numpy as np
```

```python
year = ['2021', '2022', '2023']
path = ['./강서구/', './금정구/', './기장군/', './남구/', './동구/', './동래구/', './부산진구/', './북구/', './사상구/', './사하구/', './서구/', './수영구/', './연제구/', './영도구/', './중구/', './해운대구/']
dong = ['강동동', '서제1동', '기장읍', '대연제1동', '초량제1동', '수민동', '부전제1동', '구포제1동', '삼락동', '괴정제1동', '동대신제1동', '남천제1동', '거제제1동', '남항동', '중앙동', '우제1동']

for j in range(len(path)):
    for i in range(len(year)):
        
        file = f'{dong[j]}_기온_{year[i]}01_{year[i]}12.csv'
        name = f'{dong[j]}{year[i]}'
        
        df = pd.read_csv(f'{path[j]}{file}')
        
        # 컬럼명 변경
        df.columns = ['day', 'hour', 'value']
        
        # 'Start' 문자열이 포함된 행 삭제 및 month 값 처리
        month = 1
        new_rows = []
        for _, row in df.iterrows():
            if isinstance(row['day'], str) and "Start" in row['day']:
                # Start 문자열이 포함된 행은 무시
                month += 1  # month 값을 1 증가
            else:
                row['year'] = year[i]
                row['month'] = month
                new_rows.append(row)
        
        # 새로운 DataFrame 생성
        updated_df = pd.DataFrame(new_rows)
        
        # 열 순서 정렬
        updated_df = updated_df[['year', 'month', 'day', 'hour', 'value']]
        
        # 결과 확인
        print(updated_df)
        
        # 새로운 CSV로 저장
        output_path = f"{name}.csv"
        updated_df.to_csv(f'{path[j]}{output_path}', index=False)
```
