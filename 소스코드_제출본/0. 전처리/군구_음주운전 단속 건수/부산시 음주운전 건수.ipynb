```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
data = pd.read_csv('./음주운전 건수 원본 데이터/경찰청_음주운전_2023.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
data['측정일시'] = pd.to_datetime(data['측정일시'])
```

```python
data['월별'] = data['측정일시'].dt.to_period('M')
```

```python
data.head()
```

```python
df_busan = data[data['관할경찰서'].str.startswith('부산')]
df_busan.head()
```

```python
df_busan.info()
```

```python
monthly_counts = df_busan.groupby(['관할경찰서', '월별']).size()
```

```python
pd.set_option('display.max_rows', None)
print(monthly_counts)
```
