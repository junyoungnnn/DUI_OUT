```python
import pandas as pd

# 파일 경로를 지정합니다
file_path = './음식점 원본데이터/소상공인시장진흥공단_상가(상권)정보_부산_202112.csv'
```

```python
# CSV 파일을 불러옵니다
data = pd.read_csv(file_path, encoding='utf-8', sep=',', quotechar='"')
```

```python
data.head().T
```

```python
# 공백 제거 (모든 문자열 컬럼에 적용)
data = data.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
```

```python
# 음식2021 데이터셋 필터링
excluded_categories = ['분식', '유흥주점', '음식배달서비스', '제과제빵떡케익', '커피점/카페', '기타음식업']

food2021 = data[
    (data['상권업종대분류명'] == '음식') &
    (~data['상권업종중분류명'].isin(excluded_categories))
]
```

```python
print(food2021['상권업종중분류명'].unique())
```

```python
# 필터링된 데이터 확인
print("음식2021 데이터셋:")
food2021.head().T
```

```python
# 음식2022 행정동명 개수 계산 및 정렬
food_city_counts = food2021['행정동명'].value_counts().sort_index()
```

```python
# 결과 출력
print("음식2021 행정동명 개수 (가나다순):")
print(food_city_counts)
```

```python
# 필요하면 결과를 각각 CSV로 저장
food_city_counts.to_csv('음식2021_행정동명_개수.csv', index=True, header=['Count'], encoding='utf-8-sig')

print("행정동명 개수가 음식2021_행정동명_개수.csv로 저장되었습니다.")
```
