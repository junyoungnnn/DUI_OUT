```python
import pandas as pd

# 파일 경로를 지정합니다
file_path = './음식점, 주점 원본데이터/소상공인시장진흥공단_상가(상권)정보_부산_202312.csv'
```

```python
# CSV 파일을 불러옵니다
data = pd.read_csv(file_path, encoding='utf-8')
```

```python
data.head().T
```

```python
# 공백 제거 (모든 문자열 컬럼에 적용)
data = data.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
```

```python
# 음식2023 데이터셋 필터링
food2023 = data[
    (data['상권업종대분류명'] == '음식') &
    (
        ((data['상권업종중분류명'] == '기타 간이') & (data['상권업종소분류명'].isin(['치킨', '피자']))) |
        ((data['상권업종중분류명'] == '구내식당뷔페') & (data['상권업종소분류명'] == '뷔페')) |
        (~data['상권업종중분류명'].isin(['주점', '비알코올']))
    )
]
# 주점2023 데이터셋 필터링
pub2023 = data[
    (data['상권업종대분류명'] == '음식') &
    (data['상권업종중분류명'] == '주점')
]
```

```python
print(food2023['상권업종중분류명'].unique())
```

```python
# 필터링된 데이터 확인
print("음식2023 데이터셋:")
food2023.head().T
```

```python
# 필터링된 데이터 확인
print("\n주점2023 데이터셋:")
pub2023.head()
```

```python
# 음식2023 시군구명 개수 계산 및 정렬
food_city_counts = food2023['시군구명'].value_counts().sort_index()

# 주점2023 시군구명 개수 계산 및 정렬
pub_city_counts = pub2023['시군구명'].value_counts().sort_index()
```

```python
# 결과 출력
print("음식2023 시군구명 개수 (가나다순):")
print(food_city_counts)

print("\n주점2023 시군구명 개수 (가나다순):")
print(pub_city_counts)
```

```python
# 필요하면 결과를 각각 CSV로 저장
food_city_counts.to_csv('음식2023_시군구명_개수.csv', index=True, header=['Count'], encoding='utf-8-sig')
pub_city_counts.to_csv('주점2023_시군구명_개수.csv', index=True, header=['Count'], encoding='utf-8-sig')

print("시군구명 개수가 각각 음식2023_시군구명_개수.csv와 주점2023_시군구명_개수.csv로 저장되었습니다.")
```
